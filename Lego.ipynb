```python
import pandas as pd
```

```python
df=pd.read_csv('lego_sets.csv')
df.head()
theme=pd.read_csv('parent_themes.csv')
theme.head(50)
df.head()
theme.head(50)
```

```python
merged=df.merge(theme, left_on='parent_theme', right_on='name')
merged.head()
```

```python
merged=merged.drop(columns='name_y')
#or you can you mmerged.drop(columns='name_y, inplace=TRue)
merged.head()
```

```python
merged.head()
```

```python
licensed=merged[merged['is_licensed']]
licensed.shape

```

```python
licensed=licensed.dropna(subset=['set_num'])
licensed.head()
licensed.shape
```

```python
star_wars=licensed[licensed['parent_theme']=='Star Wars']
star_wars.head()
```
